Converting 0 and 1 values in seniorcitizen column to "yes"and "no" 

```python
def convert(value):
    if value == 1:
        return"yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(convert)
df
```

```python
plt.figure(figsize = (3,4))
ax = sns.countplot( x = 'Churn' , data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of customers by churn",  fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'], labels = gb.index , autopct = "%1.2f%%")
plt.title("Percentage of customers by churn", fontsize = 10)
plt.show()
```

From above Piechart , we find out 26.54% of customers are churned out

```python
plt.figure(figsize = (3,4))
sns.countplot(x ='gender',data = df , hue = 'Churn')
plt.title("Churn based on gender", fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (3,4))
sns.countplot(x ='SeniorCitizen',data = df , hue = 'Churn')
plt.title("Churn based on gender", fontsize = 10)
plt.show()
```

```python

data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()
percentages = data.div(data.sum(axis=1), axis=0) * 100

ax = percentages.plot(kind='bar', stacked=True, figsize=(3, 4))


for bar_group in ax.containers:
    ax.bar_label(bar_group, label_type='center', fmt='%.1f%%', fontsize=8)

plt.title("Churn Based on SeniorCitizen", fontsize=10)
plt.ylabel("Percentage")
plt.xlabel("Senior Citizen")
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

From above stacked bar chart there is high percentage of senior citizens churned

```python
plt.figure(figsize = (7,8))
sns.histplot(x = "tenure", data = df, bins =72,hue = "Churn")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x ='Contract',data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn based on Contract", fontsize = 10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of columns to analyze
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots
n_cols = 3  # Number of columns in the subplot grid
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate rows needed

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))
axes = axes.flatten()  # Flatten the axes array for easy iteration

for i, column in enumerate(columns):
    sns.countplot(data=df, x=column, hue='Churn', ax=axes[i])
    axes[i].set_title(f"{column} vs Churn", fontsize=10)
    axes[i].set_xlabel(column, fontsize=8)
    axes[i].set_ylabel("Count", fontsize=8)
    axes[i].tick_params(axis='x', rotation=45)

# Remove unused subplots if the grid is larger than needed
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```


The charts reveal that services like Fiber Optic Internet and lack of Online Security or Tech Support have higher churn rates, indicating these are potential causes of churn. Services like PhoneService or MultipleLines seem to have less impact on churn, as their churn rates are relatively lower or balanced.
